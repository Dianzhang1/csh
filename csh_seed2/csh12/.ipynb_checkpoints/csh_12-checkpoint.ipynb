```python
def search_string_in_file(file_name, string_to_search):
    line_number = 0
    list_of_results = []
    with open(file_name, 'r') as read_obj:
        for line in read_obj:
            line_number += 1
            if string_to_search in line:
                list_of_results.append((line.rstrip()))
    return list_of_results
row=search_string_in_file('csh_W1/min.profile', '# TimeStep')
data= [word for line in row for word in line.split()]
data.remove("#")
data
```

```python
start_patterns = ('print "${bulkmodulus1}')
end_patterns = ('print "========================================="')

def section_with_bounds(gen):
  section_in_play = False
  for line in gen:
    if line.startswith(start_patterns):
      section_in_play = True
    if section_in_play:
      yield line
    if line.startswith(end_patterns):
      section_in_play = False
```

```python
newfile1 = open('mech_W1.txt', 'w+')
with open("csh_W1/log.lammps") as f:
  gen = section_with_bounds(f)
  for line in gen:
    newfile1.write(line)
a_file1 = open("mech_W1.txt", "r")
lines1 = a_file1.readlines()
a_file1.close()
new_file = open("mech_W1.txt", "w+")
for line in lines1:
    if line.strip("\n") != 'print "${bulkmodulus1} ${bulkmodulus2} ${Kvrh} ${shearmodulus1} ${shearmodulus2} ${Gvrh} ${E} ${poissonratio}"' :
        new_file.write(line)
new_file.close()
a_file = open("mech_W1.txt", "r")
lines1 = a_file.readlines()
a_file.close()
new_file = open("mech_W1.txt", "w+")
for line in lines1:
    if line.strip("\n") != 'print "========================================="' :
        new_file.write(line)
new_file.close()
```
